```python
import pandas as pd
import imdb
import numpy as np
from sklearn.utils import shuffle
ia = imdb.IMDb()
```

```python
movies = pd.read_csv("C:/Users/david/OneDrive/Desktop/New/Data/movie_titles.csv", encoding = "ISO-8859-1", header = None, names = ['movie_id', 'year_of_release', 'title'])
movies
```

```python
np.random.seed(76418)
movies = shuffle(movies.iloc[:4499]).reset_index(drop = True)
movies.head()
```

```python
%%time

idx = []
genres = []
for i in range(movies.shape[0]):
    try:
        mov_list = ia.search_movie(movies.loc[i, 'title'])
        count = 0
        movie = None
        for el in mov_list:
            if('year' in el.keys() and movies.loc[i, 'title'] == el['title'] and el['year'] == movies.loc[1, 'year_of_release']):
                count = count + 1
                movie = ia.get_movie(el.movieID)
        if count == 1:
            genres.append(movie['genres'][0])
            idx.append(i)
    except:
        print("excpetion on row ", i)
    if len(idx) >= 100:
        break;
idx
genres
```

```python
movie_titles = movies.iloc[idx].reset_index(drop = True)
movie_titles['genre'] = genres
movie_titles.head()
```

```python
movie_titles.to_csv("C:/Users/david/OneDrive/Desktop/New/df/sample_movie_titles.csv", index = False)
```
